```python
#1
import random
import matplotlib.pyplot as plt

total_trials = 10000
stick_win_count = 0
switch_win_count = 0

for _ in range(total_trials):
    boxes = ['10', '10', '100']
    random.shuffle(boxes)  # Randomly assign the bills to the boxes

    initial_choice = random.choice([0, 1, 2])  # Randomly select the initial choice

    if boxes[initial_choice] == '10':
        open_box = random.choice([i for i in range(3) if i != initial_choice and boxes[i] == '10'])
    else:
        open_box = random.choice([i for i in range(3) if i != initial_choice and boxes[i] == '10'])

    if boxes[initial_choice] == '100':
        stick_win_count += 1
    else:
        switch_win_count += 1

prob_stick = stick_win_count / total_trials
prob_switch = switch_win_count / total_trials

# Plotting the results
labels = ['Stick', 'Switch']
probabilities = [prob_stick, prob_switch]
colors = ['lightblue', 'lightgreen']

plt.bar(labels, probabilities, color=colors)
plt.xlabel('Strategy')
plt.ylabel('Probability of Winning')
plt.title('Monte Carlo Simulation: Stick or Switch')
plt.ylim(0, 1)
plt.show()
```
